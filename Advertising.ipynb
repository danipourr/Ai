```python
from sklearn.metrics import r2_score
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Advertising.csv')
```

```python
print(data.head())
```

```python
print(data.shape)
```

```python
print(data.dtypes)
```

```python
print(data.isnull().sum())
```

```python
data.drop(['Unnamed: 0'] , axis = 1, inplace = True)
```

```python
print(data.head())
```

```python
x = data.drop(['sales'] , axis = 1)
y = data['sales']
```

```python
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split , cross_val_score
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.20 , random_state=40)
```

```python
from sklearn.linear_model import LinearRegression , Ridge , Lasso
```

```python
regressor = LinearRegression()
```

```python
cross_sc = cross_val_score(regressor , x_train , y_train , cv=10 , scoring='r2')
```

```python
print('Cross_Score R^2 : ' , cross_sc)
```

```python
regressor.fit(x_train , y_train)
y_pred = regressor.predict((x_test))
```

```python
from sklearn.metrics import r2_score , mean_squared_error
```

```python
print('R^2 Score :' , r2_score(y_test, y_pred))
print('MSE :' , mean_squared_error(y_test, y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feature = PolynomialFeatures(degree = 2)
x_train_poly = poly_feature.fit_transform(x_train)
x_test_poly = poly_feature.transform(x_test)
```

```python
regressor_poly = LinearRegression()
regressor_poly.fit(x_train_poly , y_train)
y_predict = regressor_poly.predict(x_test_poly)
```

```python
print('R^2 Score :' , r2_score(y_test, y_predict))
print('MSE :' , mean_squared_error(y_test, y_predict))
```
